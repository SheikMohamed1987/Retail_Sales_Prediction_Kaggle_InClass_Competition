```python
#Import Neccessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

# Acquire and Preprocess dataset

```python
#==============================================================================
# import the train dataset of retail sales
#==============================================================================

df_retail_traindata = pd.read_csv ('C:\\Users\\sheikmohamed.syedtha\\Documents\\DataSets\\Train_Kaggle.csv')
```

```python
#View Head of the dataset
df_retail_traindata.head()
```

```python
#Shape of train set
df_retail_traindata.shape
```

```python
df_retail_traindata
```

```python
df_retail_traindata["Sales(In ThousandDollars)"] = df_retail_traindata.groupby(['Month','ProductCategory'])['Sales(In ThousandDollars)'].transform(lambda x: x.fillna(x.mean()))
df_retail_traindata
```

```python
#Slice the train set before imputation
X = df_retail_traindata.iloc [:,1:3].values
y = df_retail_traindata.iloc [:,3].values
```

```python
print(X,y)
```

```python
#Reshape target column to 2D array to impute
y = np.reshape(y, (-1, 1))
y
```

```python
#==============================================================================
# Encode the categorial column [Product Category]. So now instead of character values we will have
# corresponding numerical values
#==============================================================================

from sklearn.preprocessing import LabelEncoder
X_labelencoder = LabelEncoder()
X[:, 1] = X_labelencoder.fit_transform(X[:,1])
print (X)
```

```python
#Let us know look at the correlation between variables in dataset [Except Product category]
corr = df_retail_traindata.corr()
```

```python
#Seaborn heatmap to view correlations between features in dataset [Except Product category]
plt.figure(figsize=(8,6))
pltheatmap =sns.heatmap(corr)
```

```python
#Correlation between ProductCategory and sales in thousand dollars
corr1 = np.corrcoef(X[:,1].astype('float64'),y[:,0].astype('float64'))
print(corr1)

#Product Category has a strong +ve correlation with sales in thousand dollars!
```

```python
X[:,1]
```

```python
#Seaborn heatmap to view correlations between PC and Sales
plt.figure(figsize=(8,6))
pltheatmap1 =sns.heatmap(corr1)
```

```python
#PairPlot for Year, Month and sales with hue as PC
sns.pairplot(df_retail_traindata, size=4 , 
             vars=["Year","Month","Sales(In ThousandDollars)"], hue="ProductCategory")


## Pairplot clearly tells me that women's clothing category is the topmost category 
#that has higher revenue every year followed by other's and men's clothing category!!
```

```python
#PCA
from sklearn.decomposition.pca import PCA
PCA = PCA(n_components=2)
principal_components = PCA.fit_transform(X)
```

```python
principal_components
```

```python
#Cal the cumulative proportion of var explained by each component
PCA.explained_variance_ratio_
```

```python
df_X = pd.DataFrame(X)
print(df_X.columns)
df_X
```

```python
#Feature Scaling X before model training
from sklearn.preprocessing import StandardScaler
stdsclr = StandardScaler()
X_std = stdsclr.fit_transform(principal_components)
```

```python
X_std
```

# Split dataset and feature scale

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_std,y, test_size=0.16,random_state=0)
```

```python
X_train.shape
```

```python
y_train
```

# Machine Learning Model Selection and Training

```python
#Lets Train the regression model and check accuracy and RMSE
#==============================================================================
# Fitting the Polynomial Regression algorithm to the Training set
#==============================================================================

from sklearn.preprocessing import PolynomialFeatures
polyagent = PolynomialFeatures(degree=3)
X_Poly = polyagent.fit_transform(X_train) 
```

```python
X_Poly.shape
```

```python
#==============================================================================
# Fitting the Linear Regression algorithm to the Training set
#==============================================================================

from sklearn.linear_model import LinearRegression
regressoragent = LinearRegression()
regressoragent.fit (X_Poly, y_train ) 
```

```python
#Quick check accuracy of the model on train
score = regressoragent.score(X_Poly,y_train)
print(score)
```

```python
from sklearn.preprocessing import PolynomialFeatures
polyagent1 = PolynomialFeatures(degree=3)
X_Poly_test = polyagent1.fit_transform(X_test)
```

```python
#Quick check accuracy of the model on test
score = regressoragent.score(X_Poly_test,y_test)
print(score)
```

```python
predictValues = regressoragent.predict(X_Poly_test)
predictValues
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test,predictValues))
#RMSE below
```
